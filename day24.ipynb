```python
class Execution():
    def __init__(self, inst, verbose=False):
        self.x = self.y = self.z = self.w = 0
        self.instructions = inst
        self.done = False
        self.verbose = verbose

    def evaluate(self, s):
        if s == 'x':
            return self.x
        elif s == 'y':
            return self.y
        elif s == 'z':
            return self.z
        elif s == 'w':
            return self.w
        return int(s)
    
    def __str__(self):
        return f"w: {self.w}, x: {self.x}, y: {self.y}, z: {self.z}"
        
    def assign(self, s, v):
        if s == 'x':
            self.x = v
        elif s == 'y':
            self.y = v
        elif s == 'z':
            self.z = v
        elif s == 'w':
            self.w = v
        
    def execute(self, i=None):
        if len(self.instructions) == 0:
            if self.verbose: print("Done !")
            self.done = True
            return
        op, *l = self.instructions[0]
        if op == 'inp' and i is None:
            if self.verbose: print(f"Input required: ({op, *l})")
            return
        elif op == 'inp':
            var = l[0]
            self.assign(var, i)
        elif op == 'add':
            var1, var2 = l[0], l[1]
            self.assign(var1, self.evaluate(var1) + self.evaluate(var2))
        elif op == 'mul':
            var1, var2 = l[0], l[1]
            self.assign(var1, self.evaluate(var1) * self.evaluate(var2))
        elif op == 'div':
            var1, var2 = l[0], l[1]
            v1, v2 = self.evaluate(var1), self.evaluate(var2)
            self.assign(var1, int(v1 / v2))
        elif op == 'mod':
            var1, var2 = l[0], l[1]
            self.assign(var1, self.evaluate(var1) % self.evaluate(var2))
        elif op == 'eql':
            var1, var2 = l[0], l[1]
            self.assign(var1, int(self.evaluate(var1) == self.evaluate(var2)))
        self.instructions = self.instructions[1:]
        self.execute()

def check_number(s, inst):
    """
    s is a 14 digit number (as a string) with no 0
    """
    engine = Execution(inst)
    for c in s:
        engine.execute(int(c))
    assert engine.done
    if not engine.z == 0: print(engine.z)
    return engine.z

def naive_attempt(inst, n):
    solutions = []
    for _ in range(n):
        t = np.random.randint(1, 9, 14)
        s = ''.join(map(str, t))
        assert len(s) == 14
        if '0' not in s and check_number(s, inst):
            solutions.append(int(s))
    if solutions == []:
        return None
    print(solutions)
    return max(solutions)

def real_attempt(rank, inp):
    """
    constraints:
    # (i4 + 4 == i5)   
    # (i7 - 6 == i8)   **
    # (i6 - 5 == i9)   **
    # (i3 - 7 == i10)?
    # (i2 + 6 == i11)  **
    # (i12 + 1 == i13) **
    # (i1 == i14)      **
    """
    start = '93859993419899'
    assert len(start) == 14
    values = []
    for i in range(1, 10):
        s = start[:rank] + str(i) + start[rank+1:]
        values.append((s, check_number(s, inp)))
    return values
```
